### **finding the right files and filtering**

```python
print("Step 2: Filtering valid files...")
def is_valid_file_id(file_id, struct_in_path, full_out_path):
    struct_file = os.path.join(struct_in_path, f"{file_id}.npy")
    full_file = os.path.join(full_out_path, f"{file_id}.npy")
    if not os.path.exists(struct_file) or not os.path.exists(full_file):
        return False
    try:
        struct_data = np.load(struct_file)
        full_data = np.load(full_file)
        if struct_data.shape != (512, 512, 3) or full_data.shape != (512, 512, 3):
            print(f"Invalid shape for file {file_id}: struct {struct_data.shape}, full {full_data.shape}")
            return False
        return True
    except Exception as e:
        print(f"Error loading file {file_id}: {e}")
        return False

valid_file_ids_path = os.path.join(output_path, 'valid_file_ids.npy')
if os.path.exists(valid_file_ids_path):
    valid_file_ids = np.load(valid_file_ids_path).tolist()
else:
    file_ids = sorted([int(f.split('.')[0]) for f in os.listdir(train_full_out_path) if f.endswith('.npy')])
    valid_file_ids = []
    for file_id in tqdm(file_ids, desc="Checking files"):
        if is_valid_file_id(file_id, train_structure_in_path, train_full_out_path):
            valid_file_ids.append(file_id)
    np.save(valid_file_ids_path, np.array(valid_file_ids))
print(f"Number of valid files: {len(valid_file_ids)}")
print_memory_usage()
```

### **extracting features from the csv file and preproccessing the file**

```python
print("Step 3: Extracting features from CSV...")
df = pd.read_csv(excel_path)
df = df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], errors='ignore')
print(f"CSV loaded. Shape: {df.shape}")

# محاسبه مساحت از ستون geom
df['area'] = df['geom'].apply(lambda x: wkt.loads(x).area if pd.notna(x) and x != '' else 0)

# ایجاد دامی‌ها برای roomtype
roomtype_dummies = pd.get_dummies(df['roomtype'].fillna('Unknown'), prefix='room')
df = pd.concat([df, roomtype_dummies], axis=1)

# گروه‌بندی بر اساس plan_id و محاسبه فیچرها
agg_dict = {
    'area': 'sum',
    'elevation': 'mean',
    'height': 'mean',
    'floor_id': 'nunique',
    'unit_usage': 'first',
}
for col in roomtype_dummies.columns:
    agg_dict[col] = 'sum'

plan_features = df.groupby('plan_id').agg(agg_dict).reset_index()
print(f"Plan features extracted. Shape: {plan_features.shape}")

# نرمال‌سازی فیچرهای عددی
numeric_cols = ['area', 'elevation', 'height', 'floor_id'] + [col for col in plan_features.columns if col.startswith('room_')]
scaler = MinMaxScaler()
plan_features[numeric_cols] = scaler.fit_transform(plan_features[numeric_cols].fillna(0))
plan_features.to_csv(os.path.join(output_path, 'plan_features.csv'), index=False)
print("Plan features saved.")
del df, roomtype_dummies
gc.collect()
print_memory_usage()
```

### **preparing the dataset for u-net model**

```python
print("Step 4: Preparing dataset for U-Net...")
class FloorPlanDataset(Dataset):
    def __init__(self, struct_in_path, full_out_path, file_ids, augment=False):
        self.file_ids = file_ids
        self.struct_in_path = struct_in_path
        self.full_out_path = full_out_path
        self.augment = augment
        if augment:
            self.input_transform = transforms.Compose([
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.RandomVerticalFlip(p=0.5),
            ])  # حذف RandomRotation به دلیل خطا
            self.target_transform = transforms.Compose([
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.RandomVerticalFlip(p=0.5),
                transforms.Lambda(lambda x: x.long()),  # اطمینان از نوع داده Long
            ])
        else:
            self.input_transform = None
            self.target_transform = None

    def __len__(self):
        return len(self.file_ids)

    def __getitem__(self, idx):
        file_id = self.file_ids[idx]
        struct_file = os.path.join(self.struct_in_path, f"{file_id}.npy")
        full_file = os.path.join(self.full_out_path, f"{file_id}.npy")
        try:
            struct_data = np.load(struct_file).astype(np.float32)
            full_data = np.load(full_file).astype(np.float32)
            if struct_data.shape != (512, 512, 3) or full_data.shape != (512, 512, 3):
                raise ValueError(f"Invalid shape for file {file_id}")
            struct_data = (struct_data - struct_data.min()) / (struct_data.max() - struct_data.min() + 1e-8)
            struct_data = np.transpose(struct_data, (2, 0, 1))  # [C, H, W]
            full_data = full_data[:, :, 0].astype(np.int64)  # فقط کانال اول به‌عنوان ماسک
            struct_tensor = torch.tensor(struct_data)
            full_tensor = torch.tensor(full_data)
            if self.augment:
                seed = torch.randint(0, 100000, (1,)).item()
                torch.manual_seed(seed)
                struct_tensor = self.input_transform(struct_tensor)
                torch.manual_seed(seed)
                full_tensor = self.target_transform(full_tensor)
            return struct_tensor, full_tensor
        except Exception as e:
            print(f"Error loading file {file_id}: {e}")
            return torch.zeros((3, 512, 512)), torch.zeros((512, 512), dtype=torch.long)

train_dataset = FloorPlanDataset(train_structure_in_path, train_full_out_path, valid_file_ids, augment=True)
train_dataloader = DataLoader(train_dataset, batch_size=2, shuffle=True, num_workers=0, pin_memory=True)
print(f"Dataset prepared. Number of samples: {len(train_dataset)}")
print_memory_usage()
```

## **training the model**

### **define the u-net model**

```python
print("Step 5: Defining Advanced U-Net model...")
class AdvancedUNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=14):
        super(AdvancedUNet, self).__init__()

        self.enc1 = self.conv_block(in_channels, 64)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.enc2 = self.conv_block(64, 128)
        self.pool2 = nn.MaxPool2d(2, 2)
        self.enc3 = self.conv_block(128, 256)
        self.pool3 = nn.MaxPool2d(2, 2)
        self.enc4 = self.conv_block(256, 512)
        self.pool4 = nn.MaxPool2d(2, 2)

        self.middle = self.conv_block(512, 1024)

        self.upconv4 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.dec4 = self.conv_block(1024, 512)
        self.upconv3 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.dec3 = self.conv_block(512, 256)
        self.upconv2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.dec2 = self.conv_block(256, 128)
        self.upconv1 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.dec1 = self.conv_block(128, 64)

        self.final = nn.Conv2d(64, out_channels, kernel_size=1)

    def conv_block(self, in_channels, out_channels):
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3)
        )

    def forward(self, x):
        e1 = self.enc1(x)
        p1 = self.pool1(e1)
        e2 = self.enc2(p1)
        p2 = self.pool2(e2)
        e3 = self.enc3(p2)
        p3 = self.pool3(e3)
        e4 = self.enc4(p3)
        p4 = self.pool4(e4)

        m = self.middle(p4)

        u4 = self.upconv4(m)
        cat4 = torch.cat([u4, e4], dim=1)
        d4 = self.dec4(cat4)
        u3 = self.upconv3(d4)
        cat3 = torch.cat([u3, e3], dim=1)
        d3 = self.dec3(cat3)
        u2 = self.upconv2(d3)
        cat2 = torch.cat([u2, e2], dim=1)
        d2 = self.dec2(cat2)
        u1 = self.upconv1(d2)
        cat1 = torch.cat([u1, e1], dim=1)
        d1 = self.dec1(cat1)

        out = self.final(d1)
        return out

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AdvancedUNet(in_channels=3, out_channels=14).to(device)
print("Model initialized on:", device)
print_memory_usage()
```

# **settings**

```python
print("Step 6: Setting up training...")
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5)
```

### **training**

```python
# بررسی و بارگذاری آخرین وزن‌های ذخیره‌شده
start_epoch = 0
if os.path.exists(output_path):
    existing_epochs = [int(f.split('_')[2].split('.')[0]) for f in os.listdir(output_path) if f.startswith('model_epoch_') and f.endswith('.pth')]
    if existing_epochs:
        start_epoch = max(existing_epochs)
        checkpoint = torch.load(os.path.join(output_path, f'model_epoch_{start_epoch}.pth'))
        model.load_state_dict(checkpoint['model_state_dict'] if 'model_state_dict' in checkpoint else checkpoint)
        print(f"Resuming from epoch {start_epoch}")
    else:
        print("No previous checkpoints found. Starting from epoch 1.")
else:
    print("Output path does not exist. Starting from epoch 1.")

# تنظیم best_loss با مدیریت خطا
best_loss = float('inf')
if start_epoch > 0 and os.path.exists(os.path.join(output_path, f'model_epoch_{start_epoch}.pth')):
    checkpoint = torch.load(os.path.join(output_path, f'model_epoch_{start_epoch}.pth'))
    best_loss = checkpoint.get('loss', float('inf'))  # اگر 'loss' وجود نداشت، float('inf') استفاده شود

# مرحله 7: آموزش مدل با ادامه از آخرین ایپاک
print("Step 7: Starting training...")
num_epochs = 50
import time

for epoch in range(start_epoch, num_epochs):
    model.train()
    running_loss = 0.0
    valid_samples = 0
    start_time = time.time()

    for inputs, targets in tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{num_epochs}"):
        if inputs is None or targets is None:
            continue
        inputs, targets = inputs.to(device), targets.to(device)
        valid_samples += inputs.size(0)

        optimizer.zero_grad(set_to_none=True)
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        running_loss += loss.item()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    epoch_loss = running_loss / (valid_samples / 2) if valid_samples > 0 else float('inf')
    epoch_time = time.time() - start_time
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}, Time: {epoch_time:.2f} seconds")
    print(f"Learning rate: {scheduler.get_last_lr()[0]}")
    scheduler.step(epoch_loss)

    if epoch_loss < best_loss:
        best_loss = epoch_loss
        torch.save({
            'model_state_dict': model.state_dict(),
            'epoch': epoch + 1,
            'loss': epoch_loss
        }, os.path.join(output_path, 'best_unet_model.pth'))
        print("Best model saved!")

    # ذخیره وزن‌ها پس از هر ایپاک
    torch.save({
        'model_state_dict': model.state_dict(),
        'epoch': epoch + 1,
        'loss': epoch_loss
    }, os.path.join(output_path, f'model_epoch_{epoch+1}.pth'))
    print(f"Model weights saved for epoch {epoch+1}")

    # نمایش یک نمونه پیش‌بینی‌شده پس از هر ایپاک
    model.eval()
    with torch.no_grad():
        sample_input, sample_target = next(iter(train_dataloader))
        while sample_input is None or sample_target is None:
            sample_input, sample_target = next(iter(train_dataloader))
        sample_input = sample_input[0].unsqueeze(0).to(device)
        sample_target = sample_target[0].to(device)
        output = model(sample_input)
        predicted = torch.argmax(output, dim=1).cpu().numpy()[0]

        plt.figure(figsize=(15, 5))
        plt.subplot(1, 3, 1)
        plt.title("Input (Structure)")
        plt.imshow(sample_input.cpu().numpy()[0, 0, :, :], cmap='gray')
        plt.subplot(1, 3, 2)
        plt.title("Ground Truth")
        plt.imshow(sample_target.cpu().numpy(), cmap='tab20')
        plt.subplot(1, 3, 3)
        plt.title(f"Predicted (Epoch {epoch+1})")
        plt.imshow(predicted, cmap='tab20')
        plt.show()

    print_memory_usage()
    gc.collect()

print("Training completed.")
```

### **evaluating**

```python
print("Step 8: Evaluating model...")
model.load_state_dict(torch.load(os.path.join(output_path, 'best_unet_model.pth')))
model.eval()

with torch.no_grad():
    sample_input, sample_target = next(iter(train_dataloader))
    sample_input = sample_input[0].unsqueeze(0).to(device)
    sample_target = sample_target[0].to(device)
    output = model(sample_input)
    predicted = torch.argmax(output, dim=1).cpu().numpy()[0]

plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title("Input (Structure)")
plt.imshow(sample_input.cpu().numpy()[0, 0, :, :], cmap='gray')
plt.subplot(1, 3, 2)
plt.title("Ground Truth")
plt.imshow(sample_target.cpu().numpy(), cmap='tab20')
plt.subplot(1, 3, 3)
plt.title("Predicted")
plt.imshow(predicted, cmap='tab20')
plt.savefig(os.path.join(output_path, 'prediction_result.png'))
plt.show()

def calculate_iou(pred, target, num_classes=14):
    iou_per_class = []
    pred = pred.flatten()
    target = target.flatten()

    for cls in range(num_classes):
        pred_cls = (pred == cls)
        target_cls = (target == cls)
        intersection = np.logical_and(pred_cls, target_cls).sum()
        union = np.logical_or(pred_cls, target_cls).sum()
        iou_per_class.append(intersection / union if union > 0 else 1.0)

    return np.mean(iou_per_class)

iou = calculate_iou(predicted, sample_target.cpu().numpy(), num_classes=14)
print(f"Mean IoU for sample: {iou:.4f}")
print_memory_usage()
```
